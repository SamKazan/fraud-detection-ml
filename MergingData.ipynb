```python
import pandas as pd
import os
```

```python
# List all files in the directory
files = os.listdir(file_path)
file1 = 'test_identity.csv'
file2='test_transaction_merged.csv'
file3='train_identity.csv'
file4='train_transaction_merged.csv'
```

```python
# Read your files into dataframes
df1 = pd.read_csv(os.path.join(file_path, file1))
df2 = pd.read_csv(os.path.join(file_path, file2))
df3 = pd.read_csv(os.path.join(file_path, file3))
df4 = pd.read_csv(os.path.join(file_path, file4))

# Merge test files
test_set = pd.merge(df1, df2, on='TransactionID', how='inner')

# Merge train files
train_set = pd.merge(df3, df4, on='TransactionID', how='inner')
```

```python
print(df4.shape)
print(df3.shape)
```

```python
# Check the dimensions of the dataframes
print("Train Set Dimensions: ", train_set.shape)
print("Test Set Dimensions: ", test_set.shape)

# Get a quick overview of the datasets
print("\nTrain Set Overview:")
print(train_set.info())
print("\nTest Set Overview:")
print(test_set.info())

# Display the first few rows of the dataframes
print("\nTrain Set First Few Rows:")
print(train_set.head())
print("\nTest Set First Few Rows:")
print(test_set.head())

# Display the summary statistics of the dataframes
print("\nTrain Set Summary Statistics:")
print(train_set.describe())
print("\nTest Set Summary Statistics:")
print(test_set.describe())

# Check for missing values in the dataframes
print("\nTrain Set Missing Values:")
print(train_set.isnull().sum())
print("\nTest Set Missing Values:")
print(test_set.isnull().sum())
```
